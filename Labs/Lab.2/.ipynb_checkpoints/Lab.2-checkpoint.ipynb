## Responsible Use of Large Language Models (LLMs)

### Appropriate uses
- Asking for **explanations** of Python concepts (lists, loops, functions, conditionals)
- Getting **hints** or alternative approaches when you are stuck
- Debugging errors *after* you try to reason about them yourself
- Asking an LLM to **explain your own code** back to you

- Copy‑pasting complete solutions without understanding them
- Submitting code you cannot explain

*Exercise 1:* Write a function that creates an **n×n matrix** (a list of lists) representing the state of a tic‑tac‑toe game.

Use the integers:

- `0` = empty
- `1` = `"X"`
- `2` = `"O"`


````python
def tictactoe(n):
    board = []
    for i in range (n):
        row = [0] * n 
        board.append(row)
    return board
````

*Exercise 2:* Write a function that takes two integers `n` and `m` and **draws** an `n` by `m` game board.

For example, the following is a 3×3 board:

```
   --- --- --- 
  |   |   |   | 
   --- --- ---  
  |   |   |   | 
   --- --- ---  
  |   |   |   | 
   --- --- --- 
```


````python
def drawBoard(n,m):
    top = ' ---'
    bar = '|   ' 
    for i in range(n):
        print(top * m)
        print(bar * m + '|')
    print(top * m)
````

*Exercise 3:* Modify Exercise 2 so that it takes a matrix in the format from Exercise 1 and draws a tic‑tac‑toe board with `"X"`s and `"O"`s.

````python
def drawBoard(board):
    n = len(board)
    m = len(board[0])
    top = ' ---'
    for row in board:
        print(top * m)

        line = '|'
        for cell in row:
            if cell == 1:
                symbol = 'X'
            elif cell == 2:
                symbol = 'O'
            else:
                symbol = ' '

            line += f' {symbol} |'

        print(line)

    print(top * m)
````

````python
drawBoard(board)
````

*Exercise 4:* Write a function that takes an `n×n` matrix representing a tic‑tac‑toe game and returns one of the following values:

- `-1` if the game is **incomplete** (still empty spaces and no winner)
- `0` if the game is a **draw**
- `1` if **player 1** (`"X"`) has won
- `2` if **player 2** (`"O"`) has won

Here are some example inputs you can use to test your code:


````python
def game_status(board):
    n = len(board)

    for i in range(n):
        first = board[i][0]
        if first != 0:
            win = True
            for j in range(n):
                if board[i][j] != first:
                    win = False
            if win:
                return first

    for j in range(n):
        first = board[0][j]
        if first != 0:
            win = True
            for i in range(n):
                if board[i][j] != first:
                    win = False
            if win:
                return first

    first = board[0][0]
    if first != 0:
        win = True
        for i in range(n):
            if board[i][i] != first:
                win = False
        if win:
            return first

    first = board[0][n - 1]
    if first != 0:
        win = True
        for i in range(n):
            if board[i][n - 1 - i] != first:
                win = False
        if win:
            return first

    for row in board:
        for cell in row:
            if cell == 0:
                return -1  

    return 0
````

````python
winner_is_2 = [[2, 2, 0],
	[2, 1, 0],
	[2, 1, 1]]

winner_is_1 = [[1, 2, 0],
	[2, 1, 0],
	[2, 1, 1]]

winner_is_also_1 = [[0, 1, 0],
	[2, 1, 0],
	[2, 1, 1]]

no_winner = [[1, 2, 0],
	[2, 1, 0],
	[2, 1, 2]]

also_no_winner = [[1, 2, 0],
	[2, 1, 0],
	[2, 1, 0]]
````

````python
print(game_status(winner_is_2))
print(game_status(winner_is_1))
print(game_status(winner_is_also_1))
print(game_status(no_winner))
print(game_status(also_no_winner))
````

*Exercise 5:* Write a function that takes a game board, a player number, and `(row, col)` coordinates and places the correct mark (`"X"` or `"O"`) in that location.

Requirements:

- Only allow placing a mark in a previously empty location.
- Return `True` if the move was successful, and `False` otherwise.


````python
def place_move(board, player, row, col):
    if board[row][col] == 0:
        board[row][col] = player
        return True
    else:
        return False
````

````python
board = tictactoe(3)

print(place_move(board, 1, 0, 0))
print(place_move(board, 2, 0, 0))

drawBoard(board)
````

*Exercise 6:* Modify Exercise 3 to show **row and column labels** so that players can specify locations like `"A2"` or `"C1"`.

````python
def drawBoardLabeled(board):
    n = len(board)
    m = len(board[0])
    top = ' ---'

    header = '   '
    for c in range(m):
        header += f' {c+1}  '
    print(header)

    for r in range(n):
        print('  ' + top * m)
        row_label = chr(ord('A') + r)
        line = row_label + ' |'
        for c in range(m):
            cell = board[r][c]
            if cell == 1:
                symbol = 'X'
            elif cell == 2:
                symbol = 'O'
            else:
                symbol = ' '
            line += f' {symbol} |'
        print(line)
    print('  ' + top * m)
````

````python
board = tictactoe(3)

place_move(board, 1, 0, 0)
place_move(board, 2, 1, 1)
place_move(board, 1, 2, 2)

drawBoardLabeled(board)
````

*Exercise 7:* Write a function that takes a board, a player number, and a location string (as in Exercise 6), then uses your function from Exercise 5 to update the board.

````python
def place_move_string(board, player, location):
    location = location.strip().upper()

    rows = ['A','B','C','D','E']
    row = rows.index(location[0])
    col = int(location[1]) - 1

    return place_move(board, player, row, col)
````

````python
board = tictactoe(3)

print(place_move_string(board, 1, 'A1'))
print(place_move_string(board, 2, 'A1'))
print(place_move_string(board, 2, 'B2'))

drawBoardLabeled(board)
````

*Exercise 8:* Write a function that is called with a board and player number, takes input from the player using Python's `input()`, and modifies the board using your function from Exercise 7.

Keep asking for input until the player enters a valid location that results in a valid move.


````python
def player_turn(board, player):
    while True:
        location = input(f"Player {player}, your move: ")
        if place_move_string(board, player, location):
            break
        else:
            print('invalid')
````

````python
board = tictactoe(3)

drawBoardLabeled(board)
player_turn(board, 1)
drawBoardLabeled(board)
````

*Exercise 9:* Use all of the previous exercises to implement a full tic‑tac‑toe game:

- draw the board,
- repeatedly ask two players for a location,
- apply valid moves,
- check the game status until a player wins or the game is a draw.


````python
def fullGame():
    board = tictactoe(3)
    status = -1
    player = 1

    while status == -1:
        drawBoardLabeled(board)

        player_turn(board, player)
        status = game_status(board)

        if status == -1:
            if player == 1:
                player = 2
            else:
                player = 1

    # to show final board
    drawBoardLabeled(board)

    if status == 1:
        print('player 1 wins')
    elif status == 2:
        print('player 2 wins')
    else:
        print('draw')
````

````python
fullGame()
````

*Exercise 11:* Develop a version of the game where one player is the computer.

Note: you do **not** need an extensive search for the best move. For example, you can have the computer:
- block obvious losses
- otherwise try to create a winning row/column/diagonal


````python
def empty_spots(board):
    spots = []
    n = len(board)
    for r in range(n):
        for c in range(n):
            if board[r][c] == 0:
                spots.append((r, c))
    return spots

def computer_turn(board, computer_player):
    n = len(board)
    opponent = 1 if computer_player == 2 else 2

    for (r, c) in empty_spots(board):
        board[r][c] = computer_player
        if game_status(board) == computer_player:
            return True
        board[r][c] = 0 

    for (r, c) in empty_spots(board):
        board[r][c] = opponent
        if game_status(board) == opponent:
            board[r][c] = 0 
            return place_move(board, computer_player, r, c)
        board[r][c] = 0  

    spots = empty_spots(board)
    if len(spots) > 0:
        r, c = spots[0]
        return place_move(board, computer_player, r, c)

    return False
````

````python
b = tictactoe(3)
b[0][0] = 2
b[0][1] = 2
b[1][1] = 1
drawBoardLabeled(b)
computer_turn(b, 2)
print("after computer move (should win):")
drawBoardLabeled(b)
print("stats:", game_status(b))
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
def empty_spots(board):
    spots = []
    n = len(board)
    for r in range(n):
        for c in range(n):
            if board[r][c] == 0:
                spots.append((r, c))
    return spots


def evaluate_terminal(status, computer_player):
    if status == computer_player:
        return 1
    elif status == 0:
        return 0
    elif status == -1:
        return None
    else:
        return -1


def minimax_score(board, current_player, computer_player, depth, max_depth):
    status = game_status(board)
    term = evaluate_terminal(status, computer_player)
    if term is not None:
        return term
    if depth >= max_depth:
        return 0

    moves = empty_spots(board)
    if len(moves) == 0:
        return 0

    opponent = 1 if computer_player == 2 else 2

    if current_player == computer_player:
        best = -2
        for (r, c) in moves:
            board[r][c] = current_player
            score = minimax_score(board, opponent, computer_player, depth + 1, max_depth)
            board[r][c] = 0
            if score > best:
                best = score
            if best == 1:
                break
        return best
    else:
        best = 2
        for (r, c) in moves:
            board[r][c] = current_player
            score = minimax_score(board, computer_player, computer_player, depth + 1, max_depth)
            board[r][c] = 0
            if score < best:
                best = score
            if best == -1:
                break
        return best


def computer_turn_search(board, computer_player, max_depth=4):
    moves = empty_spots(board)
    if len(moves) == 0:
        return False

    opponent = 1 if computer_player == 2 else 2

    best_score = -2
    best_move = moves[0]

    for (r, c) in moves:
        board[r][c] = computer_player
        score = minimax_score(board, opponent, computer_player, 1, max_depth)
        board[r][c] = 0

        if score > best_score:
            best_score = score
            best_move = (r, c)

        if best_score == 1:
            break

    r, c = best_move
    return place_move(board, computer_player, r, c)
````

````python
def fullGame_vs_search_computer(size=3, max_depth=4, human_player=1):
    board = tictactoe(size)
    status = -1
    player = 1
    computer_player = 2 if human_player == 1 else 1

    while status == -1:
        drawBoardLabeled(board)

        if player == human_player:
            player_turn(board, human_player)
        else:
            computer_turn_search(board, computer_player, max_depth=max_depth)
            print(f"Computer (P{computer_player}) moved.")

        status = game_status(board)

        if status == -1:
            player = 2 if player == 1 else 1

    drawBoardLabeled(board)
    if status == human_player:
        print("player wins")
    elif status == computer_player:
        print("computer wins")
    else:
        print("Draw")

fullGame_vs_search_computer(size=3, max_depth=5, human_player=1)
````

*Exercise 13:* Make the 2 computer players play each-other for 10 games on a 3x3, then 4x4, then 5x5 grid. Set the max depth so that the games only take seconds. Measure the "smarter" player's win rate for each grid.

````python
def play_ai_vs_ai(size, smart_player=1, smart_depth=5, dumb_depth=2, games=10):
    wins_smart = 0
    wins_other = 0
    draws = 0

    for _ in range(games):
        board = tictactoe(size)
        status = -1
        player = 1

        while status == -1:
            if player == smart_player:
                computer_turn_search(board, player, max_depth=smart_depth)
            else:
                computer_turn_search(board, player, max_depth=dumb_depth)

            status = game_status(board)
            if status == -1:
                player = 2 if player == 1 else 1

        if status == smart_player:
            wins_smart += 1
        elif status == 0:
            draws += 1
        else:
            wins_other += 1

    return wins_smart, wins_other, draws


def experiment_10_games():
    results = {}
    configs = {
        3: (6, 2),
        4: (3, 1),
        5: (2, 1),
    }
    for size in [3, 4, 5]:
        smart_d, dumb_d = configs[size]
        w_s, w_o, d = play_ai_vs_ai(size=size, smart_player=1, smart_depth=smart_d, dumb_depth=dumb_d, games=10)
        results[size] = (w_s, w_o, d, smart_d, dumb_d)
    return results
````

````python
results = experiment_10_games()
for size in [3, 4, 5]:
    w_s, w_o, d, smart_d, dumb_d = results[size]
    print(f"{size}x{size} | smart_depth={smart_d}, other_depth={dumb_d} | smart wins={w_s}/10, other wins={w_o}/10, draws={d}")
````
